```python
import os
import pandas as pd
```

## Functions

```python
def get_returns(df, name):
    df = df.sort_values("Open Time")
    df[name] = ((df["Close"] / df["Close"].shift(1)) - 1)
    df = df.dropna(subset=[name])
    
    df["Open Time"] = pd.to_datetime(df["Open Time"])
    return df[["Open Time", name]]
```

## Daily data

```python
df_master = pd.DataFrame()
i = 0 
path = 'Crypto_data_daily/'

for file in os.listdir("Crypto_data_daily"):
    
    if file.split(".")[1] != "csv": #only read csv files and not hidden files
        continue
        
    crypto_name = file.split('.')[0]
    
    # adding header
    data = pd.read_csv(path + file).sort_values("Open Time")
    df_temp = get_returns(data, crypto_name)
    
    if i==0:
        df_master=df_temp.copy()
        i = 1
    else:
        df_master = pd.merge(df_master, df_temp, 'outer', on='Open Time')
```

```python
df_master = df_master.set_index('Open Time', drop=True)
```

```python
df_master = df_master.sort_index()
df_master.head()
```

```python
# there are a lot of nan values because cryptos dont match the same days for trading - so we simply do not have that data

# storing the data in JSON format
df_master.to_json('daily_combined.json', orient = 'split', index = 'true')
```
